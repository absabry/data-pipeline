Work with boxscores

```python
from nba_api.stats.static import teams

nba_teams = teams.get_teams()
nba_teams[0:3]
```

```python
nba_ids = [team['id'] for team in nba_teams]
nba_games_dataframe = games[games.TEAM_ID.isin(nba_ids)]
nba_games = nba_games_dataframe.GAME_ID.unique()
nba_games[0:5]
```

```python
from nba_api.stats.endpoints import boxscoretraditionalv2
game = nba_games[0]
bx = boxscoretraditionalv2.BoxScoreTraditionalV2(game_id=game).get_data_frames()[0]
```
